```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.DiGraph()
nodos=["a","b","c","d","e","f"]
G.add_nodes_from(["a","b","c","d","e","f"])
print(G.nodes())
G.add_weighted_edges_from([("a","b",8.0),("a","c",6.0),("c","d",15.0),("c","e",9.0),("e","f",13.0),("e","g",17.0),
                           ("f","g",7.0),("d","f",4.0),("d","e",14.0),("b","d",10.0)])
pos=nx.spring_layout (G)
nx.draw(G,pos, with_labels=True, node_size=1500)
nx.draw_networkx_edge_labels (G, pos, font_size=10, edge_labels=nx.get_edge_attributes (G, 'weight'))
plt.show()
try:
    print(list(nx.eulerian_path(G)))
except:
    print("no se tiene camino euleriano")
    
try:
    print(list(nx.eulerian_circuit(G)))
except:
    print("no se tiene circuito euleriano")

for x in G.nodes:
    for y in G.nodes:
        try:
            print("El camino mas corto del nodo: " + x + " al nodo: "+y+" es: " + str(list(nx.shortest_path(G,x,y))))
        except:
            print("no hay camino")
```

```python
G = nx.Graph()

G.add_edge("a", "b", weight=0.6)
G.add_edge("a", "c", weight=0.2)
G.add_edge("c", "d", weight=0.1)
G.add_edge("c", "e", weight=0.7)
G.add_edge("c", "f", weight=0.9)
G.add_edge("a", "d", weight=0.3)

elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 0.5]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] <= 0.5]

pos = nx.spring_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# edges
nx.draw_networkx_edges(G, pos, edgelist=elarge, width=6)
nx.draw_networkx_edges(
    G, pos, edgelist=esmall, width=6, alpha=0.5, edge_color="b", style="dashed"
)

# labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")

plt.axis("off")
plt.show()
```
