```python
import networkx as nx
import matplotlib.pyplot as plt
z=0
w=0
v=0
G=nx.DiGraph()
nodos=["a","b","c","d","e","f"]
G.add_nodes_from(["a","b","c","d","e","f"])
print(G.nodes())
G.add_weighted_edges_from([("a","b",8.0),("a","c",6.0),("c","d",15.0),("c","e",9.0),("e","f",13.0),("e","g",17.0),
                           ("f","g",7.0),("d","f",4.0),("d","e",14.0),("b","d",10.0)])
pos=nx.spring_layout (G)
nx.draw(G,pos, with_labels=True, node_size=1500)
nx.draw_networkx_edge_labels (G, pos, font_size=10, edge_labels=nx.get_edge_attributes (G, 'weight'))
plt.savefig("Graph.png", format="PNG")
plt.figure(figsize=(7,7))
for x in G.nodes:
    for y in G.nodes:
        if(x!=y):
            try:
                print("El camino mas corto del nodo: " + x + " al nodo: "+y+" es: " + str(list(nx.shortest_path(G,x,y))))
                v=v+1
            except:
                z=z+1
        if(x==y):
            w=w+1
print("El numero de caminos totales es de "+str(v)+", se omitieron "+str(w)+" por ser caminos al mismo vertice")
```

se resuelve usando las librerias y en mi caso use dos for para recorrer de un punto a otro, junto con la funcion nx.shortest_path para asi obtener el camino mas corto de un vertice a otro vertice, la complejidad por los for seria O(n)^2 proque iteran en la misma lista miestras que de los nodos con respecto supongo que seria algo como On oara recorrer todos
