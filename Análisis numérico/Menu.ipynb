```python
# Importacion de lbrerias 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Definicion de funciones
def my_bisection(f,a,b,tol):
    if np.sign(f(a))==np.sign(f(b)):
        raise Exception("Los escalares a y b tienen una raiz")
    m=(a+b)/2
    
    if np.abs(f(m))<tol:
        f_m=f(m)
        fig=XYplot(x,y,xmin,xmax,xlab,ylab,fig_title,symbol_color)
        fig=plt.scatter(m,f_m)
#         fig=plt.title("My bisection")
        plt.savefig("My bisection.png")
#         plt.show()
        return m
    elif np.sign(f(a))==np.sign(f(m)):
        return my_bisection(f,m,b,tol)
    elif np.sign(f(b))==np.sign(f(m)):
        return my_bisection(f,a,m,tol)

def my_newton(f,df,x0,tol):
    if abs(f(x0))<tol:
        return x0
    else:
        return my_newton(f,df,x0 - f(x0)/df(x0),tol)

    
    
def newton(f,Df,x0,epsilon,max_iter):
    xn=x0
    for n in range(0,max_iter):
        fxn=f(xn)
        if abs(fxn)<epsilon:
            print("found solution after",n,"iterations")
            fig=XYplot(x,y,xmin,xmax,xlab,ylab,fig_title,symbol_color)
            fig=plt.scatter(xn,fxn)
#             fig=plt.title("Newton")
            plt.savefig("Newton.png")

#             plt.show()
            return xn
        Dfxn=Df(xn)
        if Dfxn==0:
            print("zero derivative. No solution found.")
            return None
        xn = xn - fxn/Dfxn
    print("Exced maximum iterations. No solution found")
    return None

def XYplot(x,y,xmin,xmax,xlabb,ylabb,fig_title,symbol_color):
    fig=plt.figure()
    fig.suptitle(fig_title, fontsize=12)
    plt.plot(x,y,symbol_color)
    plt.xlim(xmin,xmax)
    plt.xlabel(xlab)
    plt.ylabel(xlab)
    plt.grid(True)
    return fig
```

```python
## Creacion de menu
ans=True
a=0
b=2
tol=0.1
max_iter=100
x0=1
f=lambda x:x**2 -2
f_prime = lambda x: 2*x
print("Aproximate solution of fx(x) vy using the following options.")
print("""
   1. Bisection
   2. Newton without iterations
   3. Newton with iterations
   4. Exit/Quit
   """)
xmin = a
xmax = b
x = np.linspace(xmin, xmax, 30)
y = f(x)
xlab = 'x'
ylab = 'y'
# fig_title = 'Root plot'
symbol_color='r-'

while ans:
    ans = input("What would you like to do? ")
    if ans=="1":
        fig_title =("My bisection")
        print("The equation is solve using: my_bisection(f,a,b,tol)")
        root= my_bisection(f,a,b,tol)
        print("root",root)
    elif ans=="2":
        print("The equation is solve using: my_newton(f,df,x0,tol)")
        root2= my_newton(f,f_prime,x0,tol)
        print("root",root2)
    elif ans=="3":
        fig_title =("Newton")
        print("The equation is solve using: newton(f,Df,x0,epsilon,max_iter)")
        root3 = newton(f, f_prime, x0, tol,max_iter)
        print("root",root3)
    elif ans=="4":
        print("\n Goodbye") 
        ans = None
    else:
        print("\n Not Valid Choice Try again")
```
