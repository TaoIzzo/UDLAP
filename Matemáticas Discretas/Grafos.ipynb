```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
nodos=["a", "b", "c", "d", "f", "g", "h", "i", "j", "k", "l", "m"]
print(nodos)
G.add_nodes_from(["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l","m"])
G.add_edges_from([("a", "b"),("a", "d"), ("a", "h"),("b", "d"),("b", "e"),("b","c"),
                 ("c","f"),("d","h"),("d","i"),("e","i"),("e","j"),("e","f"),("f","j"),
                 ("f","k"),("f","g"),("g","k"),("h","i"),("h","l"),("i","j"),("i","m"),("j","m"),
                 ("k","m"),("l","m")])
plt.figure(figsize=(7,7))
nx.draw(G, with_labels = True,font_size="20", font_color = 'black',node_size=2000,node_color="red")
grado = nx.degree_centrality(G)
# print("Centralidades de grado: ")
# for vertex in grado:
#     print(str(vertex)+": " + str(grado[vertex]))
cerca = nx.closeness_centrality(G)
# print("\nCentralidad de cercana: ")
# for vertex in cerca:
#     print(str(vertex)+": " + str(cerca[vertex]))
inter = nx.betweenness_centrality(G)
# print("\nCentralidad de intermedio: ")
# for vertex in inter:
#     print(str(vertex)+": " + str(inter[vertex]))
vector = nx.eigenvector_centrality(G)
# print("\nCentralidad de vector: ")
# for vertex in vector:
#     print(str(vertex)+": " + str(vector[vertex]))
sorted_dict = {}
sorted_keys = sorted(grado, key=grado.get)  
for x in sorted_keys:
    sorted_dict[x] = grado[x]
# print(sorted_dict)

#2 Obtenga el numero de vrtices y aristas.
vertice=G.number_of_nodes()
aristas=G.number_of_edges()
print("\nEl grafo tiene "+str(vertice)+" vertices y "+str(aristas)+" aristas")

#3 Calcule la densidad del grafo.
densidad=((2*aristas)/(vertice*(vertice-1)))
print("\nLa densidad del grafo es "+str(densidad))

#4 Calcule el dimetro del grafo.
print("\nEl diamentro del grafo es "+str(nx.diameter(G,usebounds=False))+"\n")

#5 Obtenga el grado de cada uno de los vrtices.
x=G.degree
for i in x:
    print("El vertice "+str(i[0])+" tiene de grado: "+str(i[1]))

#6 Obtenga los vrtices de grado par y de grado impar.
print(" ")
for i in x:
    if (i[1]%2==0):
        print("Es de grado par el vertice: "+str(i[0]))
    else:
        print("No es par el grado del vertice: "+str(i[0]))

# 7. Obtenga el nmero de ciclos de longitud 1 (si existen), los cuales sean del vrtice as si mismo.
ciclos=nx.cycle_basis(G)
for i in range(len(ciclos)):  
    if(len(list(ciclos)[i]))==1:
        print("El ciclo es de 1")
        print(list(nx.find_cliques(G))[i])
    else:
        print("No hay ciclos de longitud 1")
        break
# 8. Verifique si tiene un camino de Euler.
nx.has_eulerian_path(G)
if (nx.has_eulerian_path(G)==1):
    print("\nSi tiene camino de Euler")
else:
    print("\nNo tiene camino de Euler")

 # 9. Verifique si tiene un circuito de Euler.
nx.eulerian_circuit(G)
if (nx.has_eulerian_path(G)==1):
    print("\nSi tiene circuito de Euler")
else:
    print("\nNo tiene circuito de Euler")

# 10. Obtenga el numero de cliques que tiene el grafo.
cliques=nx.graph_number_of_cliques(G)
print("\nEl numero de cliques que tiene el grafo es: " +str(cliques)+"\n")

# 11. Muestre varios ejemplos (si existen) de cliques de 2, 3 y 4 vrtices.
for i in range(cliques):
    if(len(list(nx.find_cliques(G))[i]))==2:
        print("El clique es de 2")
        print(list(nx.find_cliques(G))[i])
    elif(len(list(nx.find_cliques(G))[i]))==3:
        print("El clique es de 3")
        print(list(nx.find_cliques(G))[i])
    elif(len(list(nx.find_cliques(G))[i]))==4:
        print("El clique es de 4")
        print(list(nx.find_cliques(G))[i])
    else:
        print("No se encontraron grafos de 2, 3 o 4 vertices")
# 12. Muestre un par de ejemplos de grafos isomorfos al propuesto (puede usar la librera Matplotlib).
G2=G
plt.figure(figsize=(7,7))
nx.draw(G2, with_labels = True,font_size="20", font_color = 'black',node_size=2000,node_color="blue")
vertice=G2.number_of_nodes()
aristas=G2.number_of_edges()
print("\nEl grafo tiene "+str(vertice)+" vertices y "+str(aristas)+" aristas")
x=G2.degree
for i in x:
    print("El vertice "+str(i[0])+" tiene de grado: "+str(i[1]))
densidad=((2*aristas)/(vertice*(vertice-1)))
print("\nLa densidad del grafo es "+str(densidad))
if(nx.is_isomorphic(G,G2)==1):
    print("Los grafos son isomorfos")
print("Los grafos son isomorfos debido a que es el mismo grafo pero representado de forma distinta, si se le cambia al grafo los vertices, aunque posean el mismo numero de grado, aristas y vertices dejan de ser isomorfos")
# 13. Obtenga los vrtices mas aislados en el grafo (si existen). 
sorted_dict = {}
sorted_keys = sorted(grado, key=grado.get)  
for x in sorted_keys:
    sorted_dict[x] = grado[x]
lgrado=list(sorted_dict)
if(nx.number_of_isolates(G)==0):
    print("\nNo hay vertices aisalados")
else:
    print("\nHay un vertice aisaldo")
print("Dentro del grafo, los vertices mas aislados son: "+lgrado[0]+", "+lgrado[1]+" y "+lgrado[2]+"\n")

# 14. Obtenga los vrtices mas populares, considerando la topologa del grafo. **(Centralidad)**
print(sorted_dict)
print("Las personas que deberian de sitribuir la informacion son:  "+lgrado[-1]+" y "+lgrado[-2]+"\n")

# 15. Obtenga los vrtices que pueden alcanzar a otros en el menor numero de pasos. **(Cercania)**
sorted_dict = {}
sorted_keys = sorted(cerca, key=cerca.get)  
for y in sorted_keys:
    sorted_dict[y] = cerca[y]
print(sorted_dict)
lcerca=list(sorted_dict)
print("Los vertices que pueden alcanzar a otros en menor pasos o pueden distribuir la infomracion son:  "+lcerca[-1]+" , "+lcerca[-2]+" y "+lcerca[-3]+"\n")

# 16. Obtenga aquellos vrtices que son esenciales para cualquier recorrido sobre elgrafo. **(intermediacion)**
sorted_dict = {}
sorted_keys = sorted(inter, key=inter.get)  
for y in sorted_keys:
    sorted_dict[y] = inter[y]
print(sorted_dict)
linter=list(sorted_dict)
print("Los vertices que son esenciales son:  "+linter[-1]+" y "+linter[-2] )

# 17. Obtenga el grafo egocntrico del vrtice mas popular y mustrelo de manera grafica (puede usar la librera Matplotlib).
egocentrico=(nx.ego_graph(G,lgrado[-1],radius=1, center=True, undirected=False, distance=None))
plt.figure(figsize=(10,10))
nx.draw(egocentrico,with_labels=True,node_color="yellow",node_size=2000)

# 18. Verifique si el grafo esta fuertemente conectado o no.
print("\nDebido a que el grafo es no dirigido no se puede saber si esta fuertemente concectado o no, solo se puede saber si esta conectado, es decir solo se puede ver si todos los caminos estan conectados \n")

# 19. Obtenga el camino mas corto de cada vrtice contra todos los dems. Para esta parte, proponga un nuevo grafo no-dirigido a partir del anterior que cuente con pesos en las aristas.
G3 = nx.Graph()
nodos=["a", "b", "c", "d", "f", "g", "h", "i", "j", "k", "l", "m"]
print(nodos)
G3.add_nodes_from(["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l","m"])
G3.add_weighted_edges_from([("a", "b",5),("a", "d",2), ("a", "h",4),("b", "d",7),("b", "e",8),("b","c",5),
                 ("c","f",7),("d","h",3),("d","i",1),("e","i",6),("e","j",8),("e","f",2),("f","j",4),
                 ("f","k",6),("f","g",1),("g","k",7),("h","i",8),("h","l",0),("i","j",1),("i","m",6),("j","m",9),
                 ("k","m",7),("l","m",5)])
pos=nx.spring_layout (G)
plt.figure(figsize=(10,10))
nx.draw(G3,pos, with_labels=True, node_size=1000,node_color="Green")
nx.draw_networkx_edge_labels(G3, pos, font_size=10, edge_labels=nx.get_edge_attributes (G3, 'weight'))

z=0
w=0
v=0
for x in G3.nodes:
    for y in G3.nodes:
        if(x!=y):
            try:
                print("El camino mas corto del nodo: " + x + " al nodo: "+y+" es: " + str(list(nx.shortest_path(G3,x,y))))
                v=v+1
            except:
                z=z+1
        if(x==y):
            w=w+1
print("El numero de caminos totales es de "+str(v)+", se omitieron "+str(w)+" por ser caminos al mismo vertice")
# 20. Tomando en cuenta el grafo anterior, proponga un nuevo grafo no-dirigido el cual cuente con las aristas faltantes para convertirlo en un grafo completo.
complete=nx.complete_graph(G3)
plt.figure(figsize=(7,7))
nx.draw_circular(complete, with_labels = True,font_size="20", font_color = 'black',node_size=2000,node_color="orange")
```
