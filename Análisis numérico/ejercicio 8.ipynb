```python
from numpy import *
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
# Euler's Method
def euler(t0,tn,n,y0):
   h = abs(tn-t0)/n
   t = linspace(0,tn,n+1)
   y = zeros(n+1)
   y[0] = y0
   for k in range(0,n):
      y[k+1] = y[k] + h*f(t[k],y[k]) #Forward Euler
   return y
```

```python
#Script to produce graphs
#f in the IVP y’ = f(t,y), y(t0)=y0
def f(t,y):
   return y**2-y**3
```

```python
#Runge-Kutta "Classic" Order 4 method
def RK4(t0,tn,n,y0):
   h = abs(tn-t0)/n
   t = linspace(t0,tn,n+1)
   y = zeros(n+1)
   y[0] = y0
   for i in range(0,n):
      K1 = f(t[i],y[i])
      K2 = f(t[i]+h/2,y[i]+K1*h/2)
      K3 = f(t[i]+h/2,y[i]+K2*h/2)
      K4 = f(t[i]+h,y[i]+K3*h)
      y[i+1] = y[i] + h*(K1+2*K2+2*K3+K4)/6
   return y
```

```python
#analytic solution to the IVP y’ = f(t,y), y(t0)=y0
def sol(t,t0,y0):
   C = y0*exp(-(t0-3.2)**2/2)-sin(4*t0**2)
   return exp((t-3.2)**2/2)*(sin(4*t**2) + C)
```

```python
fg =1
n = 300
t0 = 0
tn = 6
y0 = .75
t = linspace(t0,tn,n+1)
ye = euler(t0,tn,n,y0)
yrk = RK4(t0,tn,n,y0)
```

```python
# Script to produce graphs
fg = plt.figure()
plt.plot(t,yrk,'r+',label='Runge-Kutta')
plt.plot(t,ye,'bo--',label='Euler')
ysol = sol(t,t0,y0)
plt.plot(t,ysol,color ='cyan',label='Exact')
plt.grid()
fg.suptitle('Example 1', fontsize=12)
#axis([0,tn,-60,40])
plt.legend()
plt.savefig("example1.png")
plt.show()
```
