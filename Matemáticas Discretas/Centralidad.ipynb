```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
G.add_edges_from( [("John","Carla"), ("John","Simon"), ("John","Celine"), ("John","Diana"), ("Carla","Celine"), ("Carla","Diana"), 
                   ("Simon","Diana"), ("Celine", "Diana"), ("Winston", "Diana"),])
plt.figure(figsize=(6,6))
nx.draw(G, with_labels=True,font_size="20", font_color = 'blue',node_size=1000)
grado = nx.degree_centrality(G)
print("Centralidades de grado: ")
for vertex in grado:
    print(str(vertex)+": " + str(grado[vertex]))
cerca = nx.closeness_centrality(G)

sorted_dict = {}
sorted_keys = sorted(grado, key=grado.get)  
for x in sorted_keys:
    sorted_dict[x] = grado[x]
grado = nx.degree_centrality(G)
print("Centralidades de grado: ")
for vertex in grado:
    print(str(vertex)+": " + str(grado[vertex]))
cerca = nx.closeness_centrality(G)
lgrado=list(sorted_dict)
print("\n\nLas personas que deberian de sitribuir la informacion son:  "+lgrado[-1])
```
