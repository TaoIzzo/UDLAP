```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.DiGraph()
nodos=["a","b","c","d","e","f"]
G.add_nodes_from(["a","b","c","d","e","f"])
print(G.nodes())
G.add_weighted_edges_from([("a","b",8.0),("a","c",6.0),("c","d",15.0),("c","e",9.0),("e","f",13.0),("e","g",17.0),
                           ("f","g",7.0),("d","f",4.0),("d","e",14.0),("b","d",10.0)])
pos=nx.spring_layout (G)
nx.draw(G,pos, with_labels=True, node_size=1500)
nx.draw_networkx_edge_labels (G, pos, font_size=10, edge_labels=nx.get_edge_attributes (G, 'weight'))
plt.show()
try:
    print(list(nx.eulerian_path(G)))
except:
    print("no se tiene camino euleriano")
    
try:
    print(list(nx.eulerian_circuit(G)))
except:
    print("no se tiene circuito euleriano")

for x in G.nodes:
    for y in G.nodes:
        try:
            print("El camino mas corto del nodo: " + x + " al nodo: "+y+" es: " + str(list(nx.shortest_path(G,x,y))))
        except:
            print("no hay camino")
```

```python
posiciones=nx.shell_layout(G)
nx.draw_networkx_nodes(G,posiciones,nodelist=list(G.nodes()),alpha=0.5)
nx.draw_networkx_edges(G,posiciones)
plt.axis('off')
plt.show()
```

```python
G=nx.Graph()
nodos=["a","b","c","d","e"]
for vertice in nodos:
    G.add_node(vertice)
G.add_edge("a","b")
G.add_edge("b","c")
G.add_edge("b","b")
G.add_edge("c","b")
G.add_edge("c","d")
G.add_edge("d","c")
G.add_edge("d","e")
G.add_edge("d","a")
G.add_edge("e","b")
posiciones=nx.shell_layout(G)
nx.draw_networkx_nodes(G,posiciones,nodelist=list(G.nodes()),alpha=0.5)
nx.draw_networkx_edges(G,posiciones)
plt.axis('off')
plt.show()
```
