```python
import numpy as np
from sympy import *
import scipy.linalg as sc

np.seterr(invalid='ignore')
print(
    "El siguiente programa srive para resolver matrices cuadradas, favor de insertar su matriz y prrestar atencion a los mensajes")

respuesta = str(input("Agregar matriz, si/no: "))

while respuesta == "si":

    num_mat = int(input("Ingrese el numero de filas/ columnas de la matriz:"))
    print("Please Enter Elements of Matrix A:")

    Matrix_A = [[[] for i in range(num_mat)] for j in range(num_mat)]  # creating 2d list to store matrix
    for i in range(num_mat):  # setting column bounds to 3
        for j in range(num_mat):  # settting row bounds to 3
            number = float(input())
            Matrix_A[i][j] = number  # fills array using nested loops

    Matrix_B = [0 for i in range(num_mat)]
    print('Ingrese los valores del vector constante: ')
    for i in range(0, num_mat):
        Matrix_B[i] = int(input())

    print("Matrix A:")
    A = np.array(Matrix_A)
    print(Matrix_A)

    print("B:")
    B = np.array(Matrix_B)
    print(B)

    detA = np.linalg.det(Matrix_A)
    Matrix_Aa=np.array(Matrix_A)
    Matrix_Bb=np.array(Matrix_B)
    compare=(np.insert(Matrix_Aa, Matrix_Aa.shape[1],Matrix_Bb, 1))
    augmented_Matrix = Matrix(compare).applyfunc(nsimplify)
    A_red_InfinitelyManySolutions = augmented_Matrix.rref()[0]
    print("Matriz aumnetada: ",A_red_InfinitelyManySolutions)

    values_range = num_mat - 1
    values_total = num_mat * num_mat
    val_row = values_total - values_range
    sumRow=0

    for i in range(val_row, values_total):
        sumRow = sumRow + A_red_InfinitelyManySolutions[i]
    
    if round(detA, 2) != 0.00 or round(detA, 2) != -0.00:
        print('The solution is unique because its determinant is:', round(detA, 2))
    elif sumRow == 0:
        print('There is infinite solutions for this matrix')
    else:
        print('There is no solution since matrix A is singular')
    respuesta =str(input("Agregar articulo, si/no:"))
```
